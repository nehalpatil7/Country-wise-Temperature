```python
import pandas as pd

from pyecharts.charts import Map,Geo
from pyecharts import options as opts, render
from pyecharts.globals import ThemeType
```

```python
dataset=pd.read_csv("country wise temperature.csv")
Country=list(dataset['country'])
Temperature=list(dataset['2013'])
list1 = [[Country[i],Temperature[i]] for i in range(len(Country))]
```

```python
map1 = Map(init_opts=opts.InitOpts(width="1000px", height="460px",theme=ThemeType.ROMANTIC))
map1.add('Temperature in 2013', list1,maptype='world',is_map_symbol_show=False)
map1.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
```

```python
map1.set_global_opts(visualmap_opts=opts.VisualMapOpts(max_=30,is_piecewise=True,pieces=[{"min": 22},
 {"min": 18, "max": 21},
 {"min": 10, "max": 17},
 {"min": 0, "max": 9},
 {"min": -16, "max": -1},]),
 title_opts=opts.TitleOpts(title="Country wise Temperature in 2013",
 subtitle='Average in 2013',pos_left='center',padding=0,item_gap=2,
 title_textstyle_opts= opts.TextStyleOpts(color='darkblue',
 font_weight='bold',font_family='Courier New',
 font_size=30),subtitle_textstyle_opts= opts.TextStyleOpts(color='grey',
 font_weight='bold',font_family='Courier New',
 font_size=13)),legend_opts=opts.LegendOpts(is_show=False))
```

```python
map1.render_notebook()
```
